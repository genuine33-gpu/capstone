```python
# import libraries
import pandas as pd # Import Pandas for data manipulation using dataframes
import numpy as np # Import Numpy for data statistical analysis
import matplotlib.pyplot as plt # Import matplotlib for data visualisation
import seaborn as sns # Statistical data visualization
```

```python
# Import Cancer data
cancer_df = pd.read_csv('cancer.csv')
cancer_df
```

```python
#mean radius: Tümör hücrelerinin ortalama yarıçapı.
#mean texture: Tümör hücrelerinin ortalama dokusu.
#mean perimeter: Tümör hücrelerinin ortalama çevresi.
#mean area: Tümör hücrelerinin ortalama alanı.
#mean smoothness: Yüzey pürüzsüzlüğünün ortalama değeri.
#mean compactness: Kompaktlık ölçüsünün ortalama değeri.
#mean concavity: İçbükeylik özelliğinin ortalama değeri.
#mean concave points: İçbükey noktaların ortalama değeri.
#mean symmetry: Simetri ölçüsünün ortalama değeri.
#mean fractal dimension: Fraktal boyutun ortalama değeri.
```

```python
df1= cancer_df
```

```python
target_mapping = {0:  "Malignant" , 1:  "Benign" }
df1['target_mapping'] = df1['target'].map(target_mapping)
```

## CODING TASK #6. PLOT PAIRPLOT, DISPLOT, AND HEATMAPS/CORRELATIONS USING SEABORN

- **Plot the pairplot**
- **Strong correlation between the mean radius and mean perimeter, mean area and mean primeter**
- **plot the distplot**
- **Displot combines matplotlib histogram function with kdeplot() (Kernel density estimate)**
- **KDE is used to plot the Probability Density of a continuous variable.**

```python
df1.head(3)
```

```python




selected_features = ['mean radius', 'mean texture', 'mean perimeter', 'mean area', 'target_mapping']
plt.figure(figsize=(10, 19))

# Pair Plot grafiğini çizin ve renklendirmek için hue='target_mapping' kullanın
pair_plot = sns.pairplot(df1[selected_features], diag_kind='kde', hue='target_mapping')

# Legend'ı özelleştirin ve yerleşimini ayarlayın
pair_plot._legend.set_bbox_to_anchor((1.15, 1))
pair_plot._legend.set_title('Target Mapping')

plt.suptitle('Pair Plot of Selected Features')
plt.tight_layout()

plt.show()




```

```python
plt.figure(figsize=(15, 12))
# Annot parametresi ile rakamları gösteriyoruz, annot_kws ile yazı fontunu özelleştiriyoruz
sns.heatmap(df1.corr(), annot=True, cmap='coolwarm', annot_kws={'size': 8},fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python

plt.figure(figsize=(10, 6))

# "Benign" sınıfı için distplot çizin
sns.distplot(df1[df1['target_mapping'] == 'Benign']['mean area'], label='Benign', hist=False)

# "Malignant" sınıfı için distplot çizin
sns.distplot(df1[df1['target_mapping'] == 'Malignant']['mean area'], label='Malignant', hist=False)

plt.title('Distribution of Mean Area by Target')
plt.xlabel('Mean Area')
plt.legend()
plt.show()
```

```python


selected_features1 = ['mean radius', 'mean texture', 'mean perimeter', 'mean area']

# Seçilen özellikler ve target_mapping ile yeni bir veri çerçevesi oluşturun
df_selected = df1[selected_features1 + ['target_mapping']]

# Veriyi melt işlemi yaparak düzenleyin
df_melted = df_selected.melt(id_vars='target_mapping', var_name='Features', value_name='Value')

plt.figure(figsize=(12, 8))

# Boxplot grafiği çizin
sns.boxplot(data=df_melted, y='target_mapping', x='Value', hue='Features', palette='Set2')
plt.title('Boxplot of Selected Features by Target Mapping')
plt.xlabel('Target Mapping')
plt.ylabel('Value')
plt.legend(title='Features', loc='upper right')

plt.show()


```

```python



# 'target' ve 'target_mapping' sütunlarını çıkartın
selected_features2 = ['mean radius', 'mean texture', 'mean perimeter', 'mean area']

# Seçilen özelliklerle yeni bir veri çerçevesi oluşturun
df_selected = df1[selected_features2 + ['target_mapping']]

# Veriyi melt işlemi yaparak düzenleyin
df_melted = df_selected.melt(id_vars='target_mapping', var_name='Features', value_name='Value')

plt.figure(figsize=(15, 8))

# Boxplot grafiği çizin ve 'target_mapping' sınıflarına göre renklendirin
sns.boxplot(data=df_melted, y='Features', x='Value', hue='target_mapping', palette='Set2')
#plt.xticks(rotation=90)  # X eksenindeki isimleri döndür
plt.title('Boxplot of Selected Features by Target Mapping')
plt.xlabel('Value')
plt.ylabel('Features')
plt.legend(title='Target Mapping')

plt.show()

```
