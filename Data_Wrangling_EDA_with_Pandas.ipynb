# CODING TASK #7. PANDAS WITH FUNCTIONS

```python
# Let's read a CSV file using Pandas as follows
hr_df = pd.read_csv('Human_Resources.csv')
hr_df.head()
```

```python
hr_df = hr_df.dropna()
hr_df.head()
```

```python
rate = 1.10
def income_increase(hr_df):
    yeni_net_deger = hr_df * rate  # 10% increase
    return yeni_net_deger

# DataFrame'e işlevi uygulayalım
hr_df['MonthlyIncome'] = hr_df['MonthlyIncome'].apply(income_increase)
hr_df['DailyRate'] = hr_df['DailyRate'].apply(income_increase)
hr_df['HourlyRate'] = hr_df['HourlyRate'].apply(income_increase)
hr_df['MonthlyRate'] = hr_df['MonthlyRate'].apply(income_increase)
```
