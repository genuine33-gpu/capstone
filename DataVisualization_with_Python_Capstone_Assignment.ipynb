## <span style=" color:red">Determines
    
Features
    
- timestamp - timestamp field for grouping the data
- cnt - the count of a new bike shares
- t1 - real temperature in C
- t2 - temperature in C “feels like”
- hum - humidity in percentage
- wind_speed - wind speed in km/h
- weather_code - category of the weather
- is_holiday - boolean field - 1 holiday / 0 non holiday
- is_weekend - boolean field - 1 if the day is weekend
- season - category field meteorological seasons: 0-spring ; 1-summer; 2-fall; 3-winter.
    
"weather_code" category description:

- 1 = Clear ; mostly clear but have some values with haze/fog/patches of fog/ fog in vicinity
- 2 = scattered clouds / few clouds
- 3 = Broken clouds
- 4 = Cloudy
- 7 = Rain/ light Rain shower/ Light rain
- 10 = rain with thunderstorm
- 26 = snowfall
- 94 = Freezing Fog
    
Initially, the task of discovering data will be waiting for you as always. Recognize features, detect missing values, outliers etc. Review the data from various angles in different time breakdowns. <span style=" color:red">4. Plot the distribution of various discrete features on (Season, haliday, weekend and weathercode)
</span>


````python
# season - category field meteorological seasons: 0-spring ; 1-summer; 2-fall; 3-winter.
season_mapping = {0: 'Spring', 1: 'Summer', 2: 'Fall', 3: 'Winter'}
df['season'] = df['season'].map(season_mapping)


````

````python
# is_holiday - boolean field - 1 holiday / 0 non holiday
df['is_holiday'] = df['is_holiday'].map({0: 'Non-Holiday', 1: 'Holiday'})
````

````python
# is_weekend - boolean field - 1 if the day is weekend
df['is_weekend'] = df['is_weekend'].map({0: 'Weekday', 1: 'Weekend'})
````

````python
weather_mapping = {
    1: 'Clear',
    2: 'Scattered Clouds',
    3: 'Broken Clouds',
    4: 'Cloudy',
    7: 'Rain/Light Rain Shower',
    10: 'Rain with Thunderstorm',
    26: 'Snowfall',
    94: 'Freezing Fog'
}

df['weather_code'] = df['weather_code'].map(weather_mapping)
````

````python
df.head(5)
````

````python
# Plot the distribution of various discrete features on (Season, haliday, weekend and weathercode)

plt.figure(figsize=(16, 14))

plt.subplot(2, 2, 1)
ax = sns.countplot(data=df, x='season', palette='Set1')
plt.title('Season Distribution')
plt.xlabel('Season')
# label for each bar
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=12, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.subplot(2, 2, 2)
ax = sns.countplot(data=df, x='is_holiday', palette='Set2')
plt.title('Holiday Distribution')
plt.xlabel('Holiday')
# label for each bar
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=12, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.subplot(2, 2, 3)
ax = sns.countplot(data=df, x='is_weekend', palette='Set3')
plt.title('Weekend Distribution')
plt.xlabel('Weekend')
# label for each bar
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=12, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.subplot(2, 2, 4)
ax = sns.countplot(data=df, x='weather_code', palette='Set1')
plt.xticks(rotation=90)
plt.title('Weather Code Distribution')
plt.xlabel('Weather Code')
# label for each bar
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=12, color='black', xytext=(0, 5),
                textcoords='offset points')

````

<span style=" color:red">6. Make feature engineering. Extract new columns (day of the week, day of the month, hour, month, season, year etc.)
</span>

````python
df1 =df.copy()
````

````python
df.head(1)
````

````python
# Look at the data type of each variable, transform timestamp in type, and set it as index.
# Make feature engineering. Extract new columns (day of the week, day of the month, hour, month, season, year etc.)

df1['timestamp'] = pd.to_datetime(df1['timestamp'])
df1.set_index('timestamp', inplace=True)

df1['day_of_week'] = df1.index.dayofweek  # Haftanın günü (0: Pazartesi, 1: Salı, ...)
df1['day_of_month'] = df1.index.day  # Ayın günü
df1['year'] = df1.index.year  # Yıl
df1['month'] = df1.index.month  # Ay
df1['hour'] = df1.index.hour  # Saat
df1['minute'] = df1.index.minute  # Dakika
df1['second'] = df1.index.second  # Saniye




````

````python
df1.head(10)
````

````python
def map_day_of_week(day):
    day_mapping = {0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'}
    return day_mapping.get(day)

df1['day_of_week'] = df1.index.dayofweek.map(map_day_of_week)


````

````python
def map_month(month):
    month_mapping = {1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June', 7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'}
    return month_mapping.get(month)

df1['month'] = df1.index.month.map(map_month)
````

<span style=" color:red">15. Visualize the distribution of bike shares by weekday/weekend with piechart and barplot

````python
total_weekday_shares = df3[df3['is_weekend'] == "Weekday"]['cnt'].sum()
total_weekend_shares = df3[df3['is_weekend'] == "Weekend"]['cnt'].sum()

# Create a pie chart to visualize the distribution
plt.figure(figsize=(10, 5))
labels = ['Weekday', 'Weekend']
sizes = [total_weekday_shares, total_weekend_shares]
colors = ['lightblue', 'lightcoral']
explode = (0, 0.1)  # Explode the first slice (Weekday)
labeldistance = 1.1
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140, labeldistance=labeldistance)
plt.title('Distribution of Bike Shares by Weekday/Weekend')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
````
