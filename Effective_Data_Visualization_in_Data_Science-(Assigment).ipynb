## CODING TASK #6. PLOT PAIRPLOT, DISPLOT, AND HEATMAPS/CORRELATIONS USING SEABORN

- **Plot the pairplot**
- **Strong correlation between the mean radius and mean perimeter, mean area and mean primeter**
- **plot the distplot**
- **Displot combines matplotlib histogram function with kdeplot() (Kernel density estimate)**
- **KDE is used to plot the Probability Density of a continuous variable.**

```python

selected_features = ['mean radius', 'mean texture', 'mean perimeter', 'mean area', 'target_mapping']
plt.figure(figsize=(10, 19))

# Pair Plot grafiğini çizin ve renklendirmek için hue='target_mapping' kullanın
pair_plot = sns.pairplot(df1[selected_features], diag_kind='kde', hue='target_mapping')

# Legend'ı özelleştirin ve yerleşimini ayarlayın
pair_plot._legend.set_bbox_to_anchor((1.15, 1))
pair_plot._legend.set_title('Target Mapping')

plt.suptitle('Pair Plot of Selected Features')
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(15, 12))
# Annot parametresi ile rakamları gösteriyoruz, annot_kws ile yazı fontunu özelleştiriyoruz
sns.heatmap(df1.corr(), annot=True, cmap='coolwarm', annot_kws={'size': 8},fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Corr haritası çok veri içerse de bilinenin aksine işin uzmanları tarafından burdaki features toplu görülmek istenir.
# o nedenle spesifik istenilen features'lar olursa ayrı Corr haritası çizilebilir.
```

```python
#Using Matplotlib, plot the histogram for United Airlines and P&G returns using 40 bins with red color.
#Display the mean and Standard deviation for both stocks on top of the figure. What did you conclude from this graph?

# Data for UAL and PG returns
returns_ual = df_rt['UAL']
returns_pg = df_rt['PG']

# Set the number of bins and the color
n_bins = 40
color = 'red'

# Create the histograms
plt.figure(figsize=(10, 6))
plt.hist(returns_ual, bins=n_bins, color=color, alpha=0.6, label='UAL')
plt.hist(returns_pg, bins=n_bins, color=color, alpha=0.6, label='PG')

# Calculate and display mean and standard deviation
mean_ual = np.mean(returns_ual)
std_ual = np.std(returns_ual)
mean_pg = np.mean(returns_pg)
std_pg = np.std(returns_pg)

plt.axvline(mean_ual, color='blue', linestyle='dashed', linewidth=2, label=f'Mean UAL: {mean_ual:.2f}')
plt.axvline(mean_pg, color='green', linestyle='dashed', linewidth=2, label=f'Mean PG: {mean_pg:.2f}')

plt.text(mean_ual, 150, f'Std. Dev. UAL: {std_ual:.2f}', color='blue')
plt.text(mean_pg, 100, f'Std. Dev. PG: {std_pg:.2f}', color='green')

plt.title('Histogram of UAL and PG Returns')
plt.xlabel('Returns')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)

plt.show()
```
